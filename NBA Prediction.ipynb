```python
data = []
df2 = pd.DataFrame(data)
df2['player']=df1['player']
df2['pos']=df1['pos']
df2['age']=df1['age']
df2['g']=df1['g']
df2['gs']=df1['gs']
df2['pts']=df1['pts']
df2['season']=df1['season']
df2.head()
```

```python
df2=df2.drop(['player'],axis='columns')
```

```python
df2.head()
```

```python
dummies = pd.get_dummies(df2.pos)
dummies.head(3)
```

```python
df3 = pd.concat([df2,dummies],axis='columns')
df3.head()
```

```python
df4 = df3.drop('pos',axis='columns')
df4.head(2)
```

```python
y= df4.drop(['g'],axis='columns')
X= df4.g
X.head()
```

```python
y=y.drop(['season'],axis='columns')
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train.values.reshape(-1, 1),y_train)
```
